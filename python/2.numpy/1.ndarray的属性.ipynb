# ndarray的属性

### shape:   数组的形状(行数，列数，或更高维的尺寸)
### ndim:    数组的维度
### size:    总元素个数
### dtype:   元素类型(整数，浮点等)

```python
import numpy as np
# 数组的形状
arr = np.array([[1,2,3],[4,5,6]])
print('数组的形状:',arr.shape)
print('数组的维度:',arr.ndim)
print('数组的元素个数:',arr.size)
print('数组的元素类型:',arr.dtype)
print('数组的转置:',arr.T)
```
