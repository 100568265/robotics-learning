# 统计函数
### 求和，平均值，中位数，标准差，方差
### 最大值，最小值
### 分位数，累积和，累积差

```python
import numpy as np
arr = np.random.randint(1,20,8)
print(arr)
```

```python
# 求和
print(np.sum(arr))
```

```python
# 平均值
print(np.mean(arr))
```

```python
# 计算中位数
print(np.median(arr))
```

```python
# 方差
print(np.var([1,2,3]))
# 标准差
print(np.std([1,2,3]))
```
