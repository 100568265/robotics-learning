```python
# 平方根
import numpy as np
print(np.sqrt(9))
arr = np.array([1,4,9])
print(np.sqrt(arr))
```

```python
# 指数
print(np.exp(1)) # e的1次方
```

```python
# 自然对数 lnX
print(np.log(2.718281828459045))
```

```python
# sin cos
print(np.sin(np.pi/2))
print(np.cos(np.pi))
```

```python
# 绝对值
arr = np.array([1,-4,-9,16])
print(np.abs(arr))
```

```python
# 计算a的b次幂
print(np.power(arr,2))
```

```python
# 四舍五入
print(np.round(3.66))
```

```python
# 向上取整，向下取整
arr = np.array([1.6,4.7,9.8,16.9])
print(np.ceil(arr))     # 向上
print(np.floor(arr))    # 向下
```

```python
# 检测缺失值
np.isnan([1,2,3,np.nan,6,3,8])
```
