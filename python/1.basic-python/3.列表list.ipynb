### list(列表)
用方括号[]表示，是‌可变‌的有序集合，元素类型可以不同，支持增删改查等操作

```python
list1 = [1,2,3,4]
type(list1)
```

```python
# list不限制数据类型，没有长度限制
list2 = [1,'2','hahaha',3.5]
list2
```

```python
len(list2)
```

```python
# 删除一个元素
del list2[2]
list2
```

```python
# 判断是否有某个元素
a = [1,2,3,4,5,6,7,8,9]
8 in a
```

```python
# 嵌套list
b = [1,2,[3,4]]
b[2][1]
```

```python
# 计数
fruits = ['apple','banana','orange', 'apple']
fruits.count('apple')
```

### 添加元素

```python
l1 = ['banana','orange','peach']
l1.append('apple')
l1
```

```python
l1.insert(2,'watermelon')
l1
```
