### 集合set
无序、不重复的元素集合，用 {} 或 set() 创建（空集合必须用 set()）。

```python
set1 = {1,2,3,4,5}
set1
```

```python
# 并集
a = {1,2,3,4}
b = {2,3,4,5}
a.union(b)
```

```python
# 交集
a.intersection(b)
```

```python
# 差集
a.difference(b)
```

```python
# 判断子集
c = {1,2,3,4,5,6}
d = {2,3,4}
d.issubset(c)
```

```python
# 添加元素
e = {1,2,3,4,5,6}
e.add(7)
e
```
