```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy

sy.init_printing()                  # 初始化sympy的打印格式
np.set_printoptions(precision=3)    # 设置numpy数组的打印精度
np.set_printoptions(suppress=True)  # 抑制numpy的科学计数法显示
```

### 画一个网格矩阵

```python
x,y = np.arange(-3,4,1), np.arange(-3,4,1)    # 定义基础坐标点
X,Y = np.meshgrid(x,y)      # 生成网格矩阵(笛卡尔积)

fig,ax = plt.subplots(figsize=(12,7))
ax.scatter(X, Y, s=200, color='red')

# 限定坐标轴显示范围
ax.axis([-2, 2.01, -2.01, 2])

ax.spines["left"].set_position("zero")      # 把左侧的y轴（spines["left"]）移到x=0的位置（原点）
ax.spines["bottom"].set_position("zero")    # 把底部的x轴（spines["bottom"]）移到y=0的位置（原点）
ax.spines["right"].set_color("none")        # 隐藏右侧和顶部的边框线
ax.spines["top"].set_color("none")

ax.grid()
plt.show()
```

### 画一个三维空间里的平面Z=X+Y

```python
Z = X + Y
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")
ax.set_title("$z=x+y$", size=18)
#ax.scatter(X,Y,Z,s=100,label='$z=x+y$')
#ax.legend()
plt.show()
```

### 绘制三个线性方程对应的平面(三元一次方程组的几何意义)
$$
\begin{aligned}
x_1 - 2x_2 + x_3 &= 0 \\
2x_2 - 8x_3 &= 8 \\
-4x_1 - 5x_2 + 9x_3 &= -9
\end{aligned}
$$

```python
x1 = np.linspace(25,35,100)
x2 = np.linspace(10,20,20)
X1, X2 = np.meshgrid(x1, x2)

fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111, projection='3d')

X3 = 2 * X2 - X1
ax.plot_surface(X1, X2, X3, cmap="viridis", alpha=0.3)

X3 = 0.25 * X2 - 1
ax.plot_surface(X1, X2, X3, cmap="summer", alpha=0.3)

X3 = -5 / 9 * X2 + 4 / 9 * X1 - 1
ax.plot_surface(X1, X2, X3, cmap="spring", alpha=0.3)

ax.scatter(29, 16, 3, s=200, color="black")
plt.show()
```
