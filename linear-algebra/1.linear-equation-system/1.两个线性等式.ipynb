````python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy

sy.init_printing()                  # 初始化sympy的打印格式
np.set_printoptions(precision=3)    # 设置numpy数组的打印精度
np.set_printoptions(suppress=True)  # 抑制numpy的科学计数法显示
````

### 可视化：两个线性方程的系统
###### 考虑以下两个等式
```
x + y = 6
x - y = -4
```

````python
x = np.linspace(-5,5,100)   # [-5,5]这个区间，生成100个均匀分布的数值
y1 = -x + 6
y2 = x + 4

fig, ax = plt.subplots(figsize=(12,7))                      # fig:画布对象，相当于一张空白的纸    ax:坐标轴对象，相当于画纸上的坐标系，所有绘图操作都基于它
ax.scatter(1, 5, s=200, zorder=5, color='r', alpha=0.8)     # 在坐标轴上绘制散点，在(1,5)位置上画一个点

# 绘制2条方程直线
ax.plot(x, y1, lw=3, label="$x+y=6$")
ax.plot(x, y2, lw=3, label="$x-y=4$")

# 交点处画十字辅助线
ax.plot([1,1],[0,5],ls="--", color='b', alpha=0.5)          # 垂直辅助线 x坐标恒为1 y坐标为0到5
ax.plot([-5, 1], [5, 5], ls="--", color="b", alpha=0.5)     # 水平辅助线 x坐标为-5到1 y坐标恒为5

# 设置x,y轴范围
ax.set_xlim([-5, 5])
ax.set_ylim([0, 12])

ax.legend() # 图例
ax.grid()   # 网格

# 指定位置处展示交点坐标
s = '$(1,5)$'
ax.text(1, 5.5, s, fontsize=20)

# 标题
ax.set_title("Solution of $x+y=6,x-y=-4$",size=22)
````
